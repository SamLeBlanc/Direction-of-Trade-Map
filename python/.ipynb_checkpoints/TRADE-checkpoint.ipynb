```python
import numpy as np
import pandas as pd
import pickle
import country_converter as coco
import warnings
warnings.filterwarnings("ignore")
```

```python
def get_data_from_xlsx():
    df_all = pd.DataFrame()
    for i in range(1,236):
        filename = f'data/DOTS/DOTS_{i}.xlsx'
        df_country = pd.DataFrame()
        countryA = pd.read_excel(filename, index_col=0, header=2, sheet_name=0).iloc[0,0]
        print(countryA)
        for idx, direction in enumerate(['export','import']):
            trade = pd.read_excel(filename, index_col=1, header=6, sheet_name=idx)
            
            
            trade = trade.iloc[: , 1:] # drop first column
            trade = trade.iloc[1: , :] # drop first row
            
            trade_long = pd.melt(trade.reset_index(), id_vars=['index'], value_vars=trade.columns) # wide to long
            trade_long.columns = ['CountryB','Year','Value'] # rename columns
            trade_long['Direction'] = direction # create new column denoting import/export
  
            trade_long['CountryA'] = countryA # create new columns for country A

            df_country = pd.concat([df_country, trade_long]).reset_index(drop=True)
            
        df_all = pd.concat([df_all, df_country]).reset_index(drop=True) # add df_country to df_all
    return df_all
```

```python
df_all = get_data_from_xlsx()
```

```python
df_all = (df_all
            .replace('Aruba, Kingdom of the Netherlands','Aruba')
            .replace('Curaçao, Kingdom of the Netherlands','Curacao')
            .replace('Sint Maarten, Kingdom of the Netherlands','Sint Maarten'))
```

```python
df_all.CountryA = df_all.CountryA.astype(str)
df_all.CountryB = df_all.CountryB.astype(str)
df_all.Value = df_all.Value.astype(str)
df_all.Value = df_all.Value.str.replace(' e','')
df_all.Value = df_all.Value.str.replace(',','')
df_all.Value = df_all.Value.astype(float)
```

```python
df_all
```

```python
df_all.to_pickle("df_all.pkl")
```

```python
df_20 = df_all[df_all.Year == '2020']
```

```python
df_20.CountryA = coco.convert(names=df_20.CountryA, to='name_short', not_found=np.nan)
df_20.CountryB = coco.convert(names=df_20.CountryB, to='name_short', not_found=np.nan)
```

```python
df_20 = df_20.dropna()
```

```python
df_20.to_csv('2020data.csv',index=False)
```

```python
capitals = pd.read_csv('data/capitals.csv')
capitals.Country = coco.convert(names=capitals.Country, to='name_short', not_found=np.nan)
capitals = capitals.dropna()
```

```python
df = pd.merge(df_20, capitals, left_on="CountryA", right_on="Country")
df = df.rename(columns={"Capital": "CapitalA", "Cap_Lat": "CapA_Lat", "Cap_Lon": "CapA_Lon"})
df = df.drop('Country', axis=1)
df
```

```python
df = df.astype(str)
```

```python
df = pd.merge(df, capitals, left_on="CountryB", right_on="Country")
df = df.rename(columns={"Capital": "CapitalB", "Cap_Lat": "CapB_Lat", "Cap_Lon": "CapB_Lon"})
df = df.drop('Country', axis=1)
df
```

```python
df.Value = df.Value.str.replace(' e','')
df.Value = df.Value.str.replace(',','')
df.Value = df.Value.astype(float)
```

```python
df[['Value', 'CapA_Lat', 'CapB_Lat', 'CapA_Lon', 'CapB_Lon']] = df[['Value', 'CapA_Lat', 'CapB_Lat', 'CapA_Lon', 'CapB_Lon']].astype(float)
```

```python
df.to_csv('2020data.csv', index=False)
```
